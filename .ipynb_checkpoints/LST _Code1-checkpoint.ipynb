```python
from pylandtemp import ndvi
```

```python
mask = redImage == 0
ndviImage = ndvi(nirImage, redImage, mask=mask)
```

```python
plt.imshow(ndviImage)
plt.colorbar()
plt.title('{}\nRed {}'.format('NDVI', ndviImage.shape))
plt.xlabel('Column #')
plt.ylabel('Row #')
plt.show()
```
